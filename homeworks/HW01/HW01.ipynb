**модуль math**

```python
import math
# округление
a = 3.2
print(math.ceil(a), math.floor(a))
# модуль
a = -7
print(math.fabs(a))
# остаток от деления
a = 7
b = 8
print(math.fmod(a, b))
# степень
print(math.pow(a, b))
# корень квадратный
print(math.sqrt(b))
# корень кубический
print(math.pow(b, 1/3))
# экспонента
print(math.exp(b))
math.pow(math.e, b)
# логарифм
math.log(a, b)
# a=-7
# b=6
# math.log(a, b)
c=100
d=4
print(math.log2(d))
print(math.log10(c))
# тригонометрические функции
x = math.pi/3
print(math.cos(x))
print(math.sin(x))
print(math.tan(x))
# факториал
print(math.factorial(b))
#  домашнее задание: Ответьте, как сконвертировать градусы в радианы и наоборот. Сделайте это.
x=30
y=math.radians(x)
print(math.radians(x))
print(math.degrees(y))
```

**модуль NumPy**

```python
import numpy as np
# массив
a = np.array([2, 3, 4, 5])
type(a)
a.shape
# двумерный массив
b = np.array([
    [45, 67, 23, 12], 
    [56, 89, 24, 53]])
b
b.shape
# массив нулей
a=np.zeros((4, 6))
a
# массив единиц
a=np.ones((2, 4, 5))
a
# единичная
a=np.eye(6)
a
# наполненная мусором
a=np.empty((4, 4))
a
# заполнение с шагом 5
a=np.arange(20, 40, 5)
a
a=np.arange(2, 0, -0.5)
print(a)
# инструкции по выводу чисел
np.set_printoptions(threshold=1000, precision=2)


# базовые операции numpy
a=np.array([20, 30, 40, 50])
b=np.arange(4)
c = a+b
print(c)

c = a-b
print(c)

c = a*b
print(c)

c = a/b
print(c)
# бесконечность
print(np.inf+np.inf)
print(-1*np.inf)
print(np.inf*np.inf)
print(np.inf-np.inf)
# не число
print(np.nan+np.nan)
print(-1*np.nan)
print(np.nan*np.nan)
print(np.nan-np.nan)
# поэлементное возведение в степень и остаток 
print(a**b)
print(a%b)


# тригонометрический блок
print(np.cos(a))
print(np.arctan(a))
print(np.tan(a))
c=5*np.cos(a) #
c
np.round(c,2)
print(c)
np.ceil(c)
print(c)
np.fix(c)
print(c)


# агрегатные функции
a = np.array([[1, 2, 3], [4, 5, 6]])
print(np.sum(a))
print(a.sum)
print(a.min)
print(a.max)
# минимальное по столбцам
print(a.min(axis=1))
# по строкам
print(a.min(axis=0))


# срезы, индексация
a = np.arange(10) ** 4
a
print(a)
a[0]
a[2:7]
a[[7,3]]
print(a[[7,7]])
a[3:5] = 2
print(a)
print(a[::-1])
# срезы индексация двумерный массив
b=np.array([np.arange(4), 
           np.arange(4) ** 3,
           np.arange(4) * 5, 
           np.arange(4) * 11,
           np.arange(4) + 2])
print(b)
print(b[2, 3])
print(b[[2, 3], 3])
print(b[2:4,1:4])

c=b.transpose()
print(c)

print(c.reshape((5, 4)))
print(1)


```
